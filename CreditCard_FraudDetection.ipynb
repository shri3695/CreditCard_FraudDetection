**Importing Dependencies**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Unbalanced Data**

**0-real transaction
1-fraud transaction**

```python
#separating data for Analysis
real=creditcard_data[creditcard_data.Class == 0]
fraud=creditcard_data[creditcard_data.Class == 1]
```

```python
print(real.shape)
print(fraud.shape)
```

```python
real.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#comparing values
creditcard_data.groupby('Class').mean()
```

**Under Sampling**

```python
real_sample = real.sample(n=492)
```

```python
#concatenating two sample dataframes
new_dataset=pd.concat([real_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

**Splitting data into Features and targets**

```python
X= new_dataset.drop(columns='Class',axis=1)
Y= new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

**Splitting data into Training and Testing Data**

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, stratify= Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Model Training**
### Logistic Regression

```python
model=LogisticRegression(max_iter=1000)
```

```python
# training Logistic Regression Model with training data
model.fit(X_train,Y_train)
```

**Model Evaluation**

**Accuracy Score**

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy= accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data= ',training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on test data= ',test_data_accuracy)
```
